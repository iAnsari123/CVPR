Importing necessary libraries

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

Loading CIFR-10 dataset and preprocessing the data

```python
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images = train_images.reshape((60000, 28, 28, 1)).astype("float32") / 255
test_images = test_images.reshape((10000, 28, 28, 1)).astype("float32") / 255
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
list_epochs = [1, 2, 3, 4, 5]
```

Three Different Optimizers(Sgd, Adam, RMSprop) with Differen momentum

```python
# Model using SGD optimizer with momentum
sgd_model = tf.keras.Sequential(
    [
        layers.Conv2D(
            32,
            (3, 3),
            activation="relu",
            input_shape=(28, 28, 1),
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(
            64,
            (3, 3),
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(
            64,
            (3, 3),
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.Flatten(),
        layers.Dense(
            64, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.01)
        ),
        layers.Dense(
            10, activation="softmax", kernel_regularizer=tf.keras.regularizers.l2(0.01)
        ),
    ]
)

sgd_optimizer = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9, nesterov=True)
sgd_model.compile(
    optimizer=sgd_optimizer, loss="categorical_crossentropy", metrics=["accuracy"]
)
history_sgd = sgd_model.fit(
    train_images,
    train_labels,
    epochs=5,
    batch_size=128,
    validation_data=(test_images, test_labels),
)

# Model using Adam optimizer
adam_model = tf.keras.Sequential(
    [
        layers.Conv2D(
            32,
            (3, 3),
            activation="relu",
            input_shape=(28, 28, 1),
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(
            64,
            (3, 3),
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(
            64,
            (3, 3),
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.Flatten(),
        layers.Dense(
            64, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.01)
        ),
        layers.Dense(
            10, activation="softmax", kernel_regularizer=tf.keras.regularizers.l2(0.01)
        ),
    ]
)

adam_optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
adam_model.compile(
    optimizer=adam_optimizer, loss="categorical_crossentropy", metrics=["accuracy"]
)
history_adam = adam_model.fit(
    train_images,
    train_labels,
    epochs=5,
    batch_size=128,
    validation_data=(test_images, test_labels),
)

# Model using RMSprop optimizer with momentum
rmsprop_model = tf.keras.Sequential(
    [
        layers.Conv2D(
            32,
            (3, 3),
            activation="relu",
            input_shape=(28, 28, 1),
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(
            64,
            (3, 3),
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(
            64,
            (3, 3),
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.Flatten(),
        layers.Dense(
            64, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.01)
        ),
        layers.Dense(
            10, activation="softmax", kernel_regularizer=tf.keras.regularizers.l2(0.01)
        ),
    ]
)

rmsprop_optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.01, momentum=0.9)
rmsprop_model.compile(
    optimizer=rmsprop_optimizer, loss="categorical_crossentropy", metrics=["accuracy"]
)
history_rmsprop = rmsprop_model.fit(
    train_images,
    train_labels,
    epochs=5,
    batch_size=128,
    validation_data=(test_images, test_labels),
)
```

Effect of Using Regularizers (L2) in Conv2D Layers(only used for adam) 

```python
l2_model = tf.keras.Sequential(
    [
        layers.Conv2D(
            32,
            (3, 3),
            activation="relu",
            input_shape=(28, 28, 1),
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(
            64,
            (3, 3),
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(
            64,
            (3, 3),
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.Flatten(),
        layers.Dense(64, activation="relu"),
        layers.Dense(10, activation="softmax"),
    ]
)

l2_model.compile(
    optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"]
)

history_l2 = l2_model.fit(
    train_images,
    train_labels,
    epochs=5,
    batch_size=128,
    validation_data=(test_images, test_labels),
)
```

Comparison of Using Data Preprocessing vs. No Preprocessing

```python
model_no_preprocessing = tf.keras.Sequential(
    [
        layers.Conv2D(
            32,
            (3, 3),
            activation="relu",
            input_shape=(28, 28, 1),
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(
            64,
            (3, 3),
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(
            64,
            (3, 3),
            activation="relu",
            kernel_regularizer=tf.keras.regularizers.l2(0.01),
        ),
        layers.Flatten(),
        layers.Dense(
            64, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.01)
        ),
        layers.Dense(
            10, activation="softmax", kernel_regularizer=tf.keras.regularizers.l2(0.01)
        ),
    ]
)

model_no_preprocessing.compile(
    optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"]
)
history_no_preprocessing = model_no_preprocessing.fit(
    train_images,
    train_labels,
    epochs=5,
    batch_size=128,
    validation_data=(test_images, test_labels),
)
```

Plotting the results

```python
# Task 1: Plotting training accuracy for different optimizers
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(list_epochs, history_sgd.history["accuracy"], label="SGD", color="blue")
plt.plot(list_epochs, history_adam.history["accuracy"], label="Adam", color="green")
plt.plot(list_epochs, history_rmsprop.history["accuracy"], label="RMSprop", color="red")
plt.xlabel("Epochs")
plt.ylabel("Training Accuracy")
plt.title("Training Accuracy Comparison (Different Optimizers)")
plt.legend()

# Task 2: Plotting training accuracy with L2 regularization
plt.subplot(2, 2, 2)
plt.plot(
    list_epochs, history_l2.history["accuracy"], label="L2 Regularization", color="red"
)
plt.xlabel("Epochs")
plt.ylabel("Training Accuracy")
plt.title("Training Accuracy Comparison (L2 Regularization)")
plt.legend()

# Task 3: Plotting training accuracy with and without preprocessing
plt.subplot(2, 2, 3)
plt.plot(
    list_epochs,
    history_no_preprocessing.history["accuracy"],
    label="With Preprocessing",
    color="blue",
)
plt.plot(
    list_epochs,
    history_no_preprocessing.history["val_accuracy"],
    label="Without Preprocessing",
    color="red",
)
plt.xlabel("Epochs")
plt.ylabel("Training Accuracy")
plt.title("Training Accuracy Comparison (Preprocessing)")
plt.legend()

plt.tight_layout()
plt.show()
```
